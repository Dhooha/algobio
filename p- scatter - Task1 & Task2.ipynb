```r
library(ggplot2)
data <- read.csv("bl_aavh.csv")
colnames(data) <-c("aav", "pident", "length")
#head(data)

pident = data$pident
length = data$length
aav = data$aav

#qplot(Petal.Length, Petal.Width, colour = Species, shape = Predicted, data = iris)
#plot(pident, length, xlab = "Percentage of identical matches", ylab ="length of the alignment")
ggplot(data = data, aes(x = pident, y = length)) + 
geom_point(aes(color=aav))+ 
labs(x="Percentage of identical matches", y="length of the alignment")
```
